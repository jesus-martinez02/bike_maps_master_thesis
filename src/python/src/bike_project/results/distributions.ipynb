### Distribution of Distance

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

k = 1.5    
lambda_ = 7    

plt.figure(figsize=(12, 8))
plt.style.use('seaborn-v0_8')

x = np.linspace(0, 30, 1000)
y = weibull_min.pdf(x, c=k, scale=lambda_)

plt.plot(x, y, label=f'Weibull(k={k}, λ={lambda_})')
plt.xlabel('Trip Distance (km)')
plt.ylabel('PDF')
plt.title('Probability Density Function of the Weibull Distribution used for Modeling the Distance of Bicycle Trips')
plt.legend()
plt.grid(True)
plt.savefig("figures/weibull_pdf")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

k = 1.5    
lambda_ = 7    

plt.figure(figsize=(12, 8))
plt.style.use('seaborn-v0_8')

x = np.linspace(0, 30, 1000)
y = weibull_min.cdf(x, c=k, scale=lambda_)


plt.plot(x, y, label=f'Weibull(k={k}, λ={lambda_})')

percentiles = [0.25, 0.50, 0.75]
colors = ['red', 'green', 'orange']

for p, color in zip(percentiles, colors):
    q = weibull_min.ppf(p, c=k, scale=lambda_)
    plt.axvline(q, color=color, linestyle='--', alpha=0.8)
    plt.axhline(p, color=color, linestyle=':', alpha=0.5)
    plt.plot(q, p, 'o', color=color)
    plt.text(
        q + 0.3,
        p,
        f'{int(p*100)}th: {q:.2f}',
        color=color,
        fontsize=11,
        verticalalignment='bottom'
    )

plt.xlabel('Trip Distance (km)')
plt.ylabel('CDF')
plt.title(('Cumulative Distribution Function of the Weibull Distribution used for Modeling the Distance of Bicycle Trips')
)
plt.legend()
plt.grid(True)
plt.savefig("figures/weibull_cdf")
```

### Truncated normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import truncnorm

mu = 0.5        
sigma = 0.25     
lower, upper = 0, 1  

a, b = (lower - mu) / sigma, (upper - mu) / sigma

rv = truncnorm(a, b, loc=mu, scale=sigma)

x = np.linspace(lower, upper, 1000)
pdf = rv.pdf(x)

samples = rv.rvs(size=200)

plt.figure(figsize=(8,5))
plt.plot(x, pdf, 'r-', lw=2, label='Truncated Normal PDF')
plt.hist(samples, bins=20, density=True, alpha=0.3, label='Histogram from Sampled Values')
plt.title('Truncated Normal Distribution (0 to 1, μ=0.5)')
plt.xlabel('$\lambda$')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.savefig("figures/truncated_normal")
```
