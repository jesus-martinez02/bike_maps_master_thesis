### Validation with Hamburg Network

```python
import geopandas as gpd

hamburg_db_raw_gdf = gpd.GeoDataFrame(gpd.read_file("db_rad_plus_json/de_hh_up_jahr2025_EPSG_4326.json").rename(
    columns={"anzahl": "count", "geometry": "geom"}), crs=4326, geometry="geom")
hamburg_db_raw_gdf
```

```python
import os
from sqlalchemy import create_engine
db_url = (
    f"postgresql://{os.getenv('POSTGRES_USER')}:"
    f"{os.getenv('POSTGRES_PASSWORD')}@"
    f"localhost:5432/"
    f"{os.getenv('POSTGRES_DB')}"
)

engine = create_engine(db_url)
```

```python
hamburg_db_raw_gdf.to_postgis("hamburg_db_rad_2025_raw", engine, if_exists="replace")
```
